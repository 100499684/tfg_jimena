```python
import time

# Aquí puedes poner el código de monitoreo
print("📊 MONITOREO REMOTO")
print("="*40)

# Pregunta cuánto tiempo llevas
minutos = input("¿Cuántos minutos llevas de entrenamiento? (aprox): ")
epocas = input("¿Cuántas épocas se han completado?: ")

try:
    mins = float(minutos)
    eps = float(epocas)
    
    if eps > 0:
        tiempo_por_epoca = mins / eps
        print(f"\n⏱️  Tiempo por época: {tiempo_por_epoca:.1f} minutos")
        
        if tiempo_por_epoca < 30:
            print("✅ Rápido - GPU funcionando bien")
        elif tiempo_por_epoca < 60:
            print("⚠️ Normal - Aceptable")
        else:
            print("🐢 Lento - Revisar configuración")
            
        print(f"📈 Para 20 épocas total: {tiempo_por_epoca*10:.1f} minutos ({tiempo_por_epoca*20/60:.1f} horas)")
except:
    print("Por favor, introduce números válidos")
```
