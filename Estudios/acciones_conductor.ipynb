```python
import csv
from pathlib import Path
from collections import Counter

# ============================================
# CAMBIA ESTA RUTA A DONDE TENGAS EL DMD
# ============================================
DMD_PATH = "../repositorio/DMD/acciones"  # <-- CAMBIA ESTO

# Buscar todos los CSV
print("🔍 Buscando archivos CSV...")
csv_files = list(Path(DMD_PATH).rglob("*.csv"))
print(f"✅ Encontrados {len(csv_files)} archivos CSV\n")

# Contenedores para resultados
all_actions = set()
action_counts = Counter()

# Analizar cada CSV
print("📊 Analizando archivos...")
for i, csv_file in enumerate(csv_files, 1):
    if i % 10 == 0:
        print(f"  Procesados {i}/{len(csv_files)}...")
    
    try:
        with open(csv_file, 'r') as f:
            reader = csv.DictReader(f)
            
            # Leer todas las filas y contar acciones
            for row in reader:
                if 'action' in row and row['action']:
                    action = row['action'].strip()
                    if action:  # Ignorar vacíos
                        all_actions.add(action)
                        action_counts[action] += 1
    except Exception as e:
        pass

# Mostrar resultados
print(f"\n{'='*70}")
print(f"ACCIONES ENCONTRADAS ({len(all_actions)}):")
print(f"{'='*70}\n")

sorted_actions = sorted(action_counts.items(), key=lambda x: x[1], reverse=True)

for i, (action, count) in enumerate(sorted_actions, 1):
    print(f"{i:3d}. {action:50s} ({count:8d} frames)")

# Guardar en archivo
with open('acciones_DMD.txt', 'w') as f:
    f.write(f"ACCIONES DEL DATASET DMD ({len(all_actions)} únicas)\n")
    f.write("="*70 + "\n\n")
    for i, (action, count) in enumerate(sorted_actions, 1):
        f.write(f"{i:3d}. {action:50s} ({count:8d} frames)\n")

print(f"\n💾 Resultados guardados en: acciones_DMD.txt")
print("✅ ¡Completado!")
```

```python
import csv
from pathlib import Path
from collections import Counter
import os

# ============================================
# Ruta a los videos
# ============================================
DMD_PATH = "../repositorio/DMD/videos"  

# Buscar todos los videos
print("Buscando archivos mp4...")
mp4_files = list(Path(DMD_PATH).rglob("*.mp4"))

# print(mp4_files)

print(f"✅ Encontrados {len(mp4_files)} archivos mp4\n")

# Contenedores para resultados
name_body = set()
name_face = set()

# Leer el nombre de cada mp4
print("📊 Leyendo nombres...")
for i, file in enumerate(mp4_files, 1):
    if i % 10 == 0:
        print(f"  Procesados {i}/{len(mp4_files)}...")

    if ("body" in os.path.basename(file)):
        name_body.add(file)
    elif ("face" in os.path.basename(file)):
        name_face.add(file)


# Guardar en archivo
with open('names_videos.txt', 'w') as f:
    f.write(f"NOMBRES VIDEOS DEL DATASET\n")
    f.write("="*70 + "\n\n\n\n")

    # Escribimos los nombres de los videos del cuerpo
    f.write("Vídeos cuerpo\n")
    f.write("-"*20 + "\n")
    for i in sorted(name_body):
        name = os.path.basename(i)
        f.write(name + "\n")

    f.write("\n\n\n\n")
    # Escribimos los nombres de los videos de la cara
    f.write("Vídeos cara\n")
    f.write("-"*20 + "\n")
    for i in sorted(name_face):
        name = os.path.basename(i)
        f.write(name + "\n")


print(f"\n Resultados guardados en: names_videos.txt")
print("¡Completado!")
```
