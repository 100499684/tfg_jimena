```python
import os
import pandas as pd
from IPython.display import Video, display

# 1. Ruta de tu dataset en el servidor de la UC3M
ruta_base = "../repositorio/DMD/videos/"

# 2. Obtener la lista de archivos del servidor
try:
    # Leemos la carpeta y ordenamos alfabéticamente
    archivos = sorted(os.listdir(ruta_base))
except FileNotFoundError:
    print(f"Error: No se encuentra la ruta {ruta_base}. Verifica si la carpeta existe.")
    archivos = []

# 3. Lógica para filtrar y mostrar un vídeo por sujeto
sujetos_vistos = set()
conteo = 0

print("--- INICIANDO COMPROBACIÓN DE SUJETOS (TFG) ---")

for nombre in archivos:
    # Solo procesamos vídeos de cuerpo (body) para la revisión
    if "rgb_body.mp4" in nombre:
        
        # Extraemos el identificador (ej: de 'gA_1_s1...' extrae 'gA_1')
        partes = nombre.split('_')
        id_sujeto = f"{partes[0]}_{partes[1]}"
        
        # Si es la primera vez que encontramos a este sujeto, lo mostramos
        if id_sujeto not in sujetos_vistos:
            print(f"\nSujeto: {id_sujeto} | Archivo: {nombre}")
            
            # Construimos la ruta completa y mostramos el vídeo
            ruta_video = os.path.join(ruta_base, nombre)
            display(Video(ruta_video, width=320, height=240))
            
            # Marcamos como visto y sumamos al contador
            sujetos_vistos.add(id_sujeto)
            conteo += 1

print(f"\n--- Finalizado: {conteo} sujetos mostrados con éxito ---")
```
